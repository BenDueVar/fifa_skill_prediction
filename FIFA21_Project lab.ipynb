```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:/Users/benja/Desktop/Ironhack/BOOTCAMP/2.1, 2.2, 2.3 FIFA21 Project/fifa21_train.csv')
```

```python
df2 = df.drop(columns=['name','position','age','nationality','club','team & contract','height', 'weight', 'foot', 'growth',
       'joined', 'loan date end','value', 'wage', 'release clause', 'volleys', 'long shots',
       'contract', 'crossing','finishing','heading accuracy','short passing','dribbling', 'curve',
       'fk accuracy', 'long passing', 'ball control','acceleration', 'sprint speed', 'strength', 'agility', 'reactions', 'balance','ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf',
       'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm',
       'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk','gk diving', 'gk handling',
       'gk kicking', 'gk positioning', 'gk reflexes','marking', 'standing tackle',
       'sliding tackle', 'w/f', 'sm', 'a/w', 'd/w', 'ir', 'shot power', 'jumping', 'stamina','aggression', 'interceptions', 'positioning', 'vision', 'penalties',
       'composure'])
df2.columns
```

```python
nums = df2.select_dtypes(include = np.number)
cats = df2.select_dtypes(include = object)
```

```python
X = nums.loc[:,['total stats', 'base stats']]
y = nums.loc[:,['ova']]
```

```python
from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler().fit(X)
x_standardized = transformer.transform(X)
X_scaled = pd.DataFrame(x_standardized, columns=X.columns)


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.5)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
from sklearn import linear_model
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
predictions = lm.predict(X_train)
print(r2_score(y_train, predictions))
predictions_test = lm.predict(X_test)
print(r2_score(y_test, predictions_test))
```

```python

filepath = 'C:/Users/benja/Desktop/Ironhack/BOOTCAMP/2.1, 2.2, 2.3 FIFA21 Project/fifa21_train.csv'
feature_cols = ['attacking', 'skill', 'movement', 'power', 'mentality', 'defending', 'goalkeeping', 'total stats', 'base stats', 'pac', 'sho', 'pas', 'dri', 'def', 'phy', 'value']
target_col = ['ova']

def fifa_predict(path, feature_cols, target_col):
    df = pd.read_csv(path)
    df.rename(columns=str.lower, inplace=True)
    df.drop(df[df['value'] == '€0'].index, inplace=True)
    X = df.loc[:, feature_cols]
    y = df.loc[:, target_col]
    valuefix = X.loc[X['value'].str.contains('\.')]
    valuefix['value'] = valuefix['value'].str.replace('M', '00000')
    X.loc[valuefix.index, 'value'] = valuefix['value']
    X['value'] = (X['value'].str.replace('€', '').str.replace('M', 'e6').str.replace('K', 'e3').str.replace('.', '', regex=False).astype(float))
    return X


```
